```python
import cv2
import face_recognition
import os 
from google.colab.patches import cv2_imshow
```

```python
img= cv2.imread("/content/drive/MyDrive/bhargav.jpg")
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_encoding=face_recognition.face_encodings(rgb_img)[0]
```

```python
testcase= cv2.imread("/content/drive/MyDrive/testcase.jpeg")
rgb_img1=cv2.cvtColor(testcase,cv2.COLOR_BGR2RGB)
img_encoding1=face_recognition.face_encodings(rgb_img1)[0]
```

```python
(top,right,bottom,left) = face_recognition.face_locations(testcase)[0] 
print((top,right,bottom,left))
```

```python
result=face_recognition.compare_faces([img_encoding],img_encoding1)
print("this is bhargav's face:",result)
if result :
 cv2.rectangle(testcase,(left,top),(right,bottom),(255,255,100),2)
 cv2.putText(testcase,'bhargav',(left+2,bottom+20),cv2.FONT_HERSHEY_PLAIN,3, (100,0,255),2) 
 cv2_imshow(testcase) ##showing images
else :
  print("face doesn't match")
```
